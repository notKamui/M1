```python
with open("mono.txt") as f:
    mono = f.read()
    
with open("poly.txt") as f:
    poly = f.read()
```

```python
def decode(text, code, rec):
    tr = text.maketrans(code, rec)
    return text.translate(tr)
```

```python
alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

```python
ex1 = "OQFFQ CGQEFUAZ QEF FDQE RMOUXQ, XM CGQEFUAZ EGUHMZFQ QEF BXGE PURRUOUXQ"

for i in range(len(alpha)):
    print(decode(ex1, alpha[i:]+alpha[:i], alpha))
```

```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations
import random

c = Counter(ex1)
plt.bar(*zip(*c.most_common()), width=.5, color='g')
plt.show()
shift = ord('Q')-ord('E')
decode(ex1, alpha[shift:]+alpha[:shift], alpha)
```
