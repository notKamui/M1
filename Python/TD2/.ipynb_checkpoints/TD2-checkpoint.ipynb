## Exercice 1

```python
def is_palindrome(s: str) -> bool:
    return s == s[::-1]
```

```python
print(is_palindrome("abcba"))
print(is_palindrome("abb"))
```

```python
def is_pangram(s: str) -> bool:
    return set(list("abcdefghijklmnopqrstuvwxyz")).issubset(s.lower())
```

```python
print(is_pangram("The quick brown fox jumps over the lazy dog"))
print(is_pangram("This is not a pangram"))
```

```python
import functools

def remove_adjacent(l: list) -> list:
    return [] if not l else functools.reduce(lambda acc, it: acc+[it] if acc[-1] != it else acc, l, [l[0]])
```

```python
remove_adjacent([1,2,2,3,4,4,5])
```

```python
remove_adjacent([])
```

```python
def digit_sum(n: int) -> int:
    return functools.reduce(lambda acc, it: acc+int(it), list(str(n)), 0)
```

```python
digit_sum(38290) == 3 + 8 + 2 + 9 + 0
```

## Exercice 2

```python
with open(r"/usr/share/dict/french", 'r', encoding = "latin-1") as file:
    dico = [ line.strip() for line in list(file) ]
```

```python
def lookup(code: str) -> list:
    return [ word for word in dico if set(word.upper()).issubset(list(code))]
```

```python
calc_words = lookup("IZEHSGLBGDO")
scrabble_words = lookup("XWRDYTRERASZ")
```

## Exercice 3

```python
import random
import re

def obscure_text(text: str) -> str:
    segments = [ s for s in re.findall(r"\w*|\W", text) if s ]
    shuffled = []
    for sub in segments:
        if re.match(r"\w{2,}", sub):
            it = sub[0] + "".join(random.sample(list(sub)[1:-1], len(list(sub)[1:-1]))) + sub[-1]
        else:
            it = sub
        shuffled.append(it)
            
    return "".join(shuffled)
```

```python
with open("harmonie.txt", 'r', encoding = "latin-1") as file:
    print(obscure_text(file.read()))
```
